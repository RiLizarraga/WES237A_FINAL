```python
# "Data acquisition instrument IEEE 488.2 compliant" AWF generation, DAC, ADC, PWM, Digital I/O, 
from IPython.display import display, HTML; display(HTML("<style>.container { width:100% !important; }</style>"))
import threading;import time;import pynq.lib.rgbled as rgbled;from pynq.overlays.base import BaseOverlay;import asyncio
#import socket;import subprocess;import platform
import socket, subprocess, platform, math
from pynq.lib import Pmod_ADC, Pmod_DAC
base = BaseOverlay("base.bit")
from pynq.lib import Pmod_OLED

### CONST
DEBUG, OPC = True, 1
ETH_ENDL = "\r\n";Parser_Thr_flag = True 
IDN = "Data acquisition instrument IEEE 488.2 compliant, version 0.001"
State = ['INIT','OFFLINE','BUSY','EXIT'],
threads = []
STATE_MACHINE = 0b10000000#7654-3210
ACQ_WFM = 0
loggedData =[]
def setbit(n):
    global STATE_MACHINE
    STATE_MACHINE |= (1<<n)
    
def clearbit(n):
    global STATE_MACHINE
    STATE_MACHINE &= ~(1 << n) 

def testbit(n):
    global STATE_MACHINE
    return (True if (STATE_MACHINE & (1<<n)) else False)

threads_all_flag, Parser_Thr_flag, btn0_flag, btn1_flag, btn2_flag, btn3_flag = True, True, False, False, False, False    # "cond" will keep threads live or kill them all
log = []
arg1 = arg2 = arg3 = ""
led4, led5 = rgbled.RGBLED(4), rgbled.RGBLED(5)
ol = BaseOverlay("base.bit")
def leds(n, onoff):
    if n < 4:    base.leds[n].write(onoff)
    elif n == 4: led4.write( 0x2 * onoff ) # toggle and keep it green pls
    else:        led5.write( 0x2 * onoff )  
def blink(t, d, n):
    for i in range(t):       #[t]: times to blink the LED
        if n < 4:    base.leds[n].toggle()
        elif n == 4: led4.write( 0x2 * (not led4.read())) # toggle and keep it green pls
        else:        led5.write( 0x2 * (not led5.read()))

        time.sleep(d)        #[d]: duration (in seconds) for the LED to be on/off    
    if n < 4:    base.leds[n].off()
    elif n == 4: led4.write(0x0)
    else:        led5.write(0x0)     
def blinkbydelay(totalDelay, blinkingRate, n):
    times2blink = int(totalDelay/blinkingRate)
    blink(times2blink,blinkingRate,n)

def volt2Degrees(volt):
    #steinhart-hart coefficients 
    c1 = 1.129148E-3
    c2 = 234.125E-6
    c3 = 87.6741E-9
    #Volt_NTC = 1.7-0.19#1.788#2.3
    Volt_ = 3.3-volt
    Volt_NTC = float(Volt_)
    Vsupply = 3.3#3.3#5.0
    R2=10000
    NTC = R2 * (Vsupply / Volt_NTC - 1.0)# NTC in Ohm
    logNTC = math.log(NTC)
    T = 1.0 / (c1 + c2*logNTC + c3*pow(logNTC, 3));# Kelvin
    TC = T - 273.15;     # K-> C
    TF  = (9.0 / 5.0) * TC  + 32.0;     # C-> F 
    return T,TC,TF

def paddtxt(str):
    return f'{str[0:14]:<15}'

def worker_thr_parser():#[_l]: threading lock (resource), [num]: index representing the LED and thread number.
    global ACQ_WFM,STATE_MACHINE,cmd,arg1,arg2,arg3,DEBUG,OPC,log, threads_all_flag, Parser_Thr_flag, dac, adc, leds, btn0_flag, btn1_flag, btn2_flag, btn3_flag, State
    OPC = 1
    pmod_oled = Pmod_OLED(base.PMODB);pmod_oled.clear();pmod_oled.write('   R E A D Y !',0,1);
    while threads_all_flag:
        try:
            #global ACQ_WFM,STATE_MACHINE,thr_adcwf,Parser_Thr_flag
            s= socket.socket(socket.AF_INET,socket.SOCK_STREAM)
            s.bind(('',4944)) #bind to all ip in the range
            s.listen(1) #listen for 1 connection
            Parser_Thr_flag = True
            while Parser_Thr_flag:
                try:
                    connection, client_address = s.accept();
                    print ("connection from :" +str(client_address)+", "+IDN)
                    #pmod_oled.clear();pmod_oled.write('test',0,0)
                    pmod_oled.clear();pmod_oled.write('connection from:',0,0);pmod_oled.write(str(client_address[0]),0,1)
                except Exception as inst:
                    s.shutdown(socket.SHUT_RDWR)
                    s.close()
                    Parser_Thr_flag = False
                    break
                while Parser_Thr_flag:
                    try:
                        connection.send(bytes(""+ETH_ENDL, "utf-8"))#connection.send(bytes("Enter cmd :"+ETH_ENDL, "utf-8"))
                        #msg = connection.recv(8192).decode();msg = msg.upper();msg=msg[:len(msg)-2];print(msg+'<')
                        msg = connection.recv(8192).decode();msg = msg.upper();msg=msg[:len(msg)-1];print(msg+'<')
                    except Exception as inst:
                        s.shutdown(socket.SHUT_RDWR)
                        s.close()
                        Parser_Thr_flag = False
                        break
                    msg0 = msg
                    msg = msg.split()
                    cmd = arg1 = arg2 = arg3 = ''
                    if (len(msg)>0):cmd = msg[0]
                    if (len(msg)>1):arg1=msg[1]
                    if (len(msg)>2):arg2=msg[2]
                    if (len(msg)>3):arg3=msg[3]
                    if (DEBUG):print(cmd+','+arg1+','+arg2+','+arg3)
                    pmod_oled.write(f'{msg0[0:14]:<15}',0,2)#pmod_oled.write(msg0,0,2)#pmod_oled.write(cmd,0,2)
                    if cmd == "TIME?":
                        t = time.localtime();current_time = time.strftime("%H:%M:%S", t)
                        connection.send(bytes(current_time+ETH_ENDL, 'utf-8'))
                        pmod_oled.write(current_time[0:14],0,3)   
                    elif cmd == "*IDN?":
                        connection.send(bytes(IDN+ETH_ENDL, 'utf-8'))
                        #str = f'{IDN[0:14]:<15}'
                        pmod_oled.write(IDN[0:14],0,3)                        
                    elif cmd == "OS?":
                        connection.send(bytes("PYNQ Linux"+ETH_ENDL, 'utf-8'))
                        pmod_oled.write("PYNQ Linux     ",0,3)   
                    elif cmd == "*OPC?":
                        if STATE_MACHINE==0b10000000:
                            OPC=1
                        else:
                            OPC=0
                        connection.send(bytes(str(OPC)+ETH_ENDL, 'utf-8'))
                        strRaw = str(OPC)+", SM:"+str("{0:b}".format(STATE_MACHINE));pmod_oled.write(strRaw[0:14],0,3)
                        #connection.send(bytes("PYNQ Linux"+ETH_ENDL, 'utf-8'))
                        #pmod_oled.write(str(OPC)+"             ",0,3) 
                    elif cmd == "DAC":#DAC 1.2
                        dac = Pmod_DAC(ol.PMODA)
                        dac.write(float(arg1))
                        pmod_oled.write('OK            ',0,3)
                        connection.send(bytes("OK"+ETH_ENDL, 'utf-8'))
                        del dac
                    elif cmd == "ADC0?":#read(ch1=1, ch2=0, ch3=0)
                        adc = Pmod_ADC(ol.PMODA)
                        value = adc.read(ch1=1,ch2=1,ch3=1)
                        connection.send(bytes(str(value)+ETH_ENDL, 'utf-8'))
                        #strRaw = str(value);pmod_oled.write(strRaw[0:14],0,3)
                        strRaw = str(value);pmod_oled.write(f'{strRaw[0:14]:<15}',0,3)
                        del adc
                    elif cmd == "ADC1?":
                        adc = Pmod_ADC(ol.PMODA)
                        value = adc.read(ch1=1,ch2=0,ch3=0)
                        connection.send(bytes(str(value[0])+ETH_ENDL, 'utf-8'))
                        strRaw = str(value[0]);pmod_oled.write(f'{strRaw[0:14]:<15}',0,3)
                        #strRaw = str(value[0]);pmod_oled.write(strRaw[0:14],0,3)                        
                        del adc
                    elif cmd == "ADC2?":
                        adc = Pmod_ADC(ol.PMODA)
                        value = adc.read(ch1=0,ch2=1,ch3=0)
                        connection.send(bytes(str(value[0])+ETH_ENDL, 'utf-8'))
                        strRaw = str(value[0]);pmod_oled.write(f'{strRaw[0:14]:<15}',0,3)
                        #strRaw = str(value[0]);pmod_oled.write(strRaw[0:14],0,3)
                        del adc
                    elif cmd == "ADC3?":
                        adc = Pmod_ADC(ol.PMODA)
                        value = adc.read(ch1=0,ch2=0,ch3=1)
                        connection.send(bytes(str(value[0])+ETH_ENDL, 'utf-8'))
                        strRaw = str(value[0]);pmod_oled.write(f'{strRaw[0:14]:<15}',0,3)
                        #strRaw = str(value[0]);pmod_oled.write(strRaw[0:14],0,3)
                        del adc
                    elif cmd == "THR1K?":#Kelvin
                        adc = Pmod_ADC(ol.PMODA)
                        value = adc.read(ch1=0,ch2=0,ch3=1)
                        kelvin = volt2Degrees(value[0])
                        connection.send(bytes(str(kelvin[0])+ETH_ENDL, 'utf-8'))
                        strRaw = str(kelvin[0]);pmod_oled.write(strRaw[0:14],0,3)
                        del adc       
                    elif cmd == "THR1C?":#Celcius
                        adc = Pmod_ADC(ol.PMODA)
                        value = adc.read(ch1=0,ch2=0,ch3=1)
                        kelvin = volt2Degrees(value[0])
                        connection.send(bytes(str(kelvin[1])+ETH_ENDL, 'utf-8'))
                        strRaw = str(kelvin[1]);pmod_oled.write(strRaw[0:14],0,3)
                        del adc         
                    elif cmd == "THR1F?":#Far
                        adc = Pmod_ADC(ol.PMODA)
                        value = adc.read(ch1=0,ch2=0,ch3=1)
                        kelvin = volt2Degrees(value[0])
                        connection.send(bytes(str(kelvin[2])+ETH_ENDL, 'utf-8'))
                        strRaw = str(kelvin[2]);pmod_oled.write(strRaw[0:14],0,3)
                        del adc                                 
                    elif cmd == "ADCWF":#ADCWF 100 3
                        setbit(0)
                        connection.send(bytes("OK"+ETH_ENDL, 'utf-8'))  
                        strRaw = "SM: 0b"+str("{0:b}".format(STATE_MACHINE));pmod_oled.write(strRaw[0:14],0,3)
                        
                    elif cmd == "EXIT" or Parser_Thr_flag == False or threads_all_flag == False:
                        #connection.send(bytes("Adios"+ETH_ENDL, 'utf-8'))
                        s.shutdown(socket.SHUT_RDWR)
                        #s.close()
                        break                        
                    else:
                        try:
                            connection.send(bytes("error: "+cmd+ " invalid command"+ETH_ENDL, 'utf-8'))
                        except Exception as inst:
                            s.shutdown(socket.SHUT_RDWR)
                            s.close()
                            Parser_Thr_flag = False
                            break
        finally:
            s.close()
        time.sleep(2)

def worker_thr_dac(freq,acq):#[_l]: threading lock (resource), [num]: index representing the LED and thread number.
    import time
    global loggedData,ACQ_WFM,STATE_MACHINE,cmd,arg1,arg2,arg3,thr_adcwf,log,OPC,DEBUG,threads_all_flag, Parser_Thr_flag, dac, adc, leds, btn0_flag, btn1_flag, btn2_flag, btn3_flag, State
    
    while threads_all_flag:#if DEBUG:print('freq='+freq+","+"Acq="+acq)
        if testbit(ACQ_WFM):
            print('-', end='')
            ### acq
            freq = int(arg1)
            acq_time = int(arg2)
            log_interval_us = 0#int(arg3)
            print("Starting Acquisition waveform @freq="+str(freq)+", acq_time="+str(acq_time)+", interval_uS="+str(log_interval_us))
            loggedData =[]
            period = 1/freq
            adc = Pmod_ADC(ol.PMODA)
            adc.start_log(1,0,0,log_interval_us)# Assume Channel 0 is connected to the waveform generator
            time.sleep(acq_time*period)
            print("Stopped")
            loggedData = adc.get_log();print(loggedData)
            clearbit(ACQ_WFM)
            # Draw the figure
            %matplotlib inline
            import matplotlib.pyplot as plt
            plt.plot(range(len(loggedData)), loggedData, 'ro')
            plt.title('PMOD ADC Waveform')
            plt.axis([0, len(loggedData), min(loggedData), max(loggedData)])
            plt.show()
        else:
            time.sleep(2)
            print('.', end='')
         
        time.sleep(2)
    ###OPC = 1
    #< > *opc? CHECK FOR ALL BITS CLEARED    
        
async def get_btns(_loop):
    btns = base.btns_gpio
    global STATE_MACHINE,DEBUG,threads_all_flag, Parser_Thr_flag, leds, btn0_flag, btn1_flag, btn2_flag, btn3_flag, State
    while threads_all_flag:
        await asyncio.sleep(0.1)
        if btns.read() != 0:
            if (btns[0].read()==1):
                await asyncio.sleep(0.3)
                print("button 0 pressed")
                time.sleep(0.1)
                btn0_flag = True
            elif (btns[1].read()==1):                
                await asyncio.sleep(0.3)
                print("button 1 pressed")        
                btn0_flag = True
            elif (btns[2].read()==1):
                await asyncio.sleep(0.3)
                print("button 2 pressed")              
                btn0_flag = True
            else:
                await asyncio.sleep(0.3)
                btn3_flag = True
                print("button 3 pressed, Exit program")
                Parser_Thr_flag = False
                threads_all_flag = False
                await asyncio.sleep(1)
                _loop.stop()
'''                                    Start here                                    '''
print("*** Listening service READY ***\n\n")

### Critical Section threading locks to use
criticalsection_1 = threading.Lock()
criticalsection_2 = threading.Lock()

### Create needed Threads (don't start them up yet) 
thr1 = threading.Thread(target=worker_thr_parser, args=())
threads.append(thr1)
thr_adcwf = threading.Thread(target=worker_thr_dac, args=(arg1,arg2))
threads.append(thr_adcwf)
### Start Threads    
for thr in threads:
    thr.start()

loop = asyncio.new_event_loop();  # Instance event_loop object
loop.create_task(get_btns(loop)); # take user input buttons    

loop.run_forever()
loop.close()
### Stop Live Threads, after  commanded by user buttons
for thr in threads:
    thr.join()
    print('{} joined'.format(t.name))

for n in range(5):
    leds(n,0)
print("Done.")     
```
